## Método de Primitivos



````python
def qiskit_sampler(circuito_com_medidas, shots = None, backend = None):
    '''
    Método para executar o Sampler
    >>> O Sampler realiza medidas em circuitos quânticos e retorna a contagem de qubits.
        A partir dos resultados do Sampler, podemos estimar as probabilidade de cada resultado.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        backend  : backend para rodar o Sampler

    Saída:
        resultado do Sampler

    Observação: os resultados do Sampler são rotulados pelo nomes dos registros clássicos.
                Para um circuito com medidas definidas por 'measure_all', o rótulo é 'meas'.
                Para circuitos com medidas definidas de início sem denominação explícita,
                os rótulos são 'c', 'c1', 'c2', etc.

    Tutorial do Sampler: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/sampler-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import SamplerV2 as Sampler

    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    from qiskit.transpiler import generate_preset_pass_manager
    pm = generate_preset_pass_manager(optimization_level=2, backend = backend)

    # Transpilação

    isa_circuitos = pm.run(circuito_com_medidas)

    if shots is None:
        shots = 2**10
    else:
        shots = shots

    sampler = Sampler(backend)
    job = sampler.run([(isa_circuitos, )], shots = shots)
    result = job.result()

    # Identificação do nome do primeiro registro clássico
    classical_register_name = isa_circuitos.cregs[0].name
    counts = result[0].data[classical_register_name].get_counts()

    return counts, result

def qiskit_estimator(circuito_sem_medidas, hamiltoniano, precisao = None, backend = None):
    '''
    Método para executar o Estimator
    >>> O Estimator estima valores esperados de circuitos e observáveis.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        hamiltoniano (SparsePauliOp): hamiltoniano a ser medido
        precisao (float): precisão do resultado
        backend  : backend para rodar o Estimator

    Saída:
        resultado do Estimator

    Tutorial do EstimatorV2: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/estimator-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import EstimatorV2 as Estimator
    from qiskit.transpiler import generate_preset_pass_manager

    # Use the existing service object
    # service = QiskitRuntimeService()
    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    if precisao is None:
        precisao = 2**(-10)
    else:
        precisao = precisao

    # Transpilação

    pm = generate_preset_pass_manager(backend = backend, optimization_level=1)
    isa_circuito = pm.run(circuito_sem_medidas)
    isa_hamiltoniano = hamiltoniano.apply_layout(isa_circuito.layout)

    # Estimator:

    estimator = Estimator(mode=backend)

    # Job
    job = estimator.run([(isa_circuito, isa_hamiltoniano, )], precision = precisao)
    pub_result = job.result()[0]

    return pub_result.data.evs
````

````python
# Métodos

import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
import qiskit.quantum_info as qi
from qiskit.circuit.library import QFT, QFTGate, HamiltonianGate


def quantiza_vetor(b_vetor = list):
    '''
    Algoritmo que codifica um vetor b em amplitude, retornando o circuito que prepara o estado |b > normalizado

    Entrada:
        b_vetor (list): vetor

    Saída:
        qc (QuantumCircuit): circuito quântico que prepara o estado |b > normalizado
    '''

    # Número de qubits para representar o vetor

    num_qubits = int(np.ceil(np.log2(len(b_vetor))))

    # Normalização e completamento do vetor até a coordenada 2**n

    vetor_normalizado = b_vetor/np.linalg.norm(b_vetor)
    vetor_normalizado_completo = np.zeros(2**num_qubits)
    vetor_normalizado_completo[:len(vetor_normalizado)] = vetor_normalizado

    # Circuito de inicialização

    qc = QuantumCircuit(num_qubits)
    qc.initialize(vetor_normalizado_completo, range(num_qubits))

    return qc

def quantiza_unitario(unitario = list):
    '''
    Algoritmo que quantiza um operador unitário U,
    retornando um circuito que implementa U

    Entrada:
        U (list): operador unitário

    Saída:
        qc (QuantumCircuit): circuito quântico que implementa U
    '''

    dimensao = len(unitario)

    num_qubits = int(np.ceil(np.log2(dimensao)))

    unitario2 = np.identity(2**num_qubits, dtype=complex)

    for i in range(dimensao):
        for j in range(dimensao):
            unitario2[i][j] = unitario[i][j]

    gate = qi.Operator(unitario2)

    qc = QuantumCircuit(num_qubits)

    qc.unitary(gate, range(num_qubits), label = "U")

    return qc

def quantiza_expoperador(operador_hermitiano = list, tempo = float):
    '''
    Algoritmo que quantiza um operador hermitiano A em dado tempo t,
    retornando o circuito que prepara o unitário exp(-2.pi.i.A.t)

    Entrada:
        operador_hermitiano (list): operador
        time (float)..............: tempo

    Saída:
        qc (QuantumCircuit): circuito quântico que prepara o unitário  exp(-2.pi.i.A.t)
    '''

    dimensao = len(operador_hermitiano)

    num_qubits = int(np.ceil(np.log2(dimensao)))

    hamiltoniano = np.identity(2**num_qubits, dtype = complex)

    for i in range(dimensao):
        for j in range(dimensao):
            hamiltoniano[i][j] = operador_hermitiano[i][j]

    # Cria a porta exp(-2.pi.i.A.t)
    time = 2*np.pi*tempo
    ham_gate = HamiltonianGate(hamiltoniano, time, label="exp(-2.pi.H.t)")
    qc = QuantumCircuit(num_qubits)
    qc.append(ham_gate, range(num_qubits))

    return qc

def Crot(num_qubits = int, t0_inverso = None):

    from qiskit.circuit.library import RYGate #qiskit.circuit.library.standard_gates

    if t0_inverso is None:
        t0_inverso = 2**num_qubits

    # Circuito com n qubits de controle e um qubit auxiliar alvo
    qc = QuantumCircuit(num_qubits + 1, name="Crot")

    # Rotações controladas

    for k in range(1,2**num_qubits):

        # angulo
        bin_string = bin(k)[2:].zfill(num_qubits)
        bin_string_invertida = bin_string[::-1]
        numero = int(bin_string, 2)/2**num_qubits
        seno_semiangulo = 1/(numero*t0_inverso)
        angulo = 2*np.arcsin(seno_semiangulo)

        # rotação controlada

        ry_gate= RYGate(angulo)
        crot = ry_gate.control(num_qubits, ctrl_state = bin_string_invertida)

        # ação

        qc.append(crot, list(range(num_qubits + 1)))

    return qc

def EstimativaFase(qc_unitario, qc_autoestado, num_qubits = int, num_qubits_aux = int, shots = None, backend = None, medir = False):
    '''
    Função que implementa o algoritmo para Estimativa de Fase.

    Inputs:
        qc_unitario (QuantumCircuit)..: circuito quântico que que implementa o operador unitário
        qc_autoestado (QuantumCircuit): circuito quântico que prepara o auto-estado do operador unitário
        num_qubits (int)..........: número de qubits do circuito principal
        num_qubits_aux (int)......: número de qubits do circuito auxiliar
        shots (int)...............: número de medições (do registro auxiliar)
        backend...................: dispositivo para realização dos experimentos quânticos

    Output:
        fase (float): estimativa de fase

    Métodos requeridos:
        QuantumCircuit, QuantumRegister, ClassicalRegister
        qiskit.circuit.library.QFT
        qiskit_ibm_runtime.SamplerV2
    '''

    if shots == None:
        shots = 2**12

    if backend == None:
        backend = FakeBrisbane()

    qr = QuantumRegister(num_qubits)
    qr_aux = QuantumRegister(num_qubits_aux)
    cr_aux = ClassicalRegister(num_qubits_aux)

    # Cria o circuito quântico
    if medir is False:
        qc = QuantumCircuit(qr, qr_aux)
    elif medir is True:
        qc = QuantumCircuit(qr, qr_aux, cr_aux)

    # Preparação do auto-estado
    qc.append(qc_autoestado, list(qr))

    # Aplicação de Hadamard nos qubits auxiliares
    for k in range(num_qubits_aux):
        qc.h(qr_aux[k])

    # Convert qc_unitario to a gate once for efficiency
    unitario_as_gate = quantiza_unitario.to_gate()

    # Aplicação das potências controladas do operador unitário

    # U^(2^0) = U:
    ctrl_gate_k0 = unitario_as_gate.control(1)

    # Aplicação de U^(2^k)-controlado
    qc.append(ctrl_gate_k0, [qr_aux[0]] + list(qr))

    # U^(2^k), para k = 1,..., num_qubits_aux -1:
    for k in range(1, num_qubits_aux):
        # Circuito que representa U^(2^k)
        qc_unitario_potencia = QuantumCircuit(num_qubits)
        for _ in range(2**k):
            qc_unitario_potencia.append(unitario_as_gate, list(range(num_qubits)))

        # Circuito de U^(2^k)-controlado
        ctrl_gate_potencia = qc_unitario_potencia.control(1)

        # Aplicação de U^(2^k)-controlado
        qc.append(ctrl_gate_potencia, [qr_aux[k]] + list(qr))

    # Aplicação da TFQ-inversa nos qubits auxiliares
    qc.append(QFT(num_qubits = num_qubits_aux, approximation_degree=0, do_swaps = True, inverse = True, insert_barriers = False, name= "QFTinv"), list(qr_aux))

    if medir is False:
        return qc

    elif medir is True:

       # Medida dos qubits auxiliares
        qc.measure(qr_aux, cr_aux)

        # Execução do algoritmo pelo Sampler
        #medida = qiskit_sampler(qc, shots = shots, backend = backend)

        if backend is None:
            from qiskit_ibm_runtime.fake_provider import FakeBrisbane
            backend = FakeBrisbane()
        else:
            backend = backend

        from qiskit.transpiler import generate_preset_pass_manager
        pm = generate_preset_pass_manager(optimization_level=2, backend = backend)

        # Transpilação

        isa_circuitos = pm.run(qc)

        if shots is None:
           shots = 2**12
        else:
            shots = shots

        # Simulação

        sampler = Sampler(backend)
        job = sampler.run([(isa_circuitos, )], shots = shots)
        result = job.result()

        # Contagens

        classical_register_name = isa_circuitos.cregs[0].name # Identificação do nome do primeiro registro clássico
        counts = result[0].data[classical_register_name].get_counts()

        # Identificação da fase (estado mais provável da base computacional)

        fase = next(iter(counts)) # Primeira chave do dicionário counts
        for key in counts.keys():
            if counts[key] > counts[fase]:
                fase = key

        return fase, counts
````

````python
def Emax(operador = list):
    '''
    Determina o máximo dos módulos das entradas do operador
    '''
    emax = np.abs(operador[0][0])
    for linha in operador:
        for entrada in linha:
            abs_entrada = np.abs(entrada)
            if abs_entrada > emax:
                emax = abs_entrada

    return emax

def Emin(operador = list):
    '''
    Determina o menor dos módulos das entradas não nulas do operador
    '''
    emin = np.abs(operador[0][0])
    for linha in operador:
        for entrada in linha:
            abs_entrada = np.abs(entrada)
            if abs_entrada >0 and abs_entrada < emin:
                emin = abs_entrada

    return emin


def HHL(operador_hermitiano = list, vetor = list, t0 = None, t1 = None, qubits_vetor = int, qubits_fase = int, shots = None, backend = None):
    '''
    Algoritmo HHL para gerar o estado quantizado da solução do sistema
    de equações lineares Ax = b, onde
    - b é um vetor e
    - A é um operador hermitiano inversível.

    Entradas:
        qubits_vetor (int).: número de qubits do registrar o vetor quantizado
        qubits_fase (int)..: número de qubits para registrar fases
        vetor (list)...: vetor |b > do registro quântico principal
        operador_hermitiano (list): operador hermitiano inversível

    Saída:
        qc_hhl (QuantumCircuit): circuito quântico que implementa o algoritmo HHL
        ........................ gerando o circuito que prepara o estado |x >
        ........................ que corresponde à quantização do vetor A^{-1}b.

    '''

    qr_vetor = QuantumRegister(qubits_vetor)
    qr_fase = QuantumRegister(qubits_fase)
    qr_aux = QuantumRegister(1)
    cbit = ClassicalRegister(1)
    qbits = qubits_vetor + qubits_fase + 1
    qc = QuantumCircuit(qr_vetor, qr_fase, qr_aux, cbit)

    # Inicialização

    qc_vetor = quantiza_vetor(vetor) # Circuito que prepara o estado quântico do vetor

    qc.append(qc_vetor, range(qubits_vetor))

    # Estimativa de fase

    if t1 is None:
        t1 = Emax(operador_hermitiano)

    if t0 is None:
        t0 = Emin(operador_hermitiano)/t1

    qc_unitario = quantiza_expoperador(operador_hermitiano, tempo = 1/t1) # Circuito que implementa a exponencial do operador hermitiano

    qc_fase = EstimativaFase(qc_unitario, qc_vetor, num_qubits = qubits_vetor, qubits_aux = qubits_fase, shots = None, backend = None, medir = False)

    qc.append(qc_fase, range(qubits_vetor + qubits_fase))

    # Rotações controladas

    qc_rotacoes = Crot(qubits_fase, t0_inverso = 1/t0)

    qc.append(qc_rotacoes, range(qbits)[qubits_vetor:qbits])

    # Estimativa de fase inversa

    qc_unitario_inverso = quantiza_expoperador(operador_hermitiano, tempo = - 1/t1) # Circuito que implementa a exponencial do operador hermitiano

    qc_fase_inversa = EstimativaFase(qc_unitario_inverso, qc_vetor, qubits_vetor, qubits_fase, shots = None, backend = None, medir = False)

    qc.append(qc_fase_inversa, range(qubits_vetor + num_qubits_fase))

    # # Medida

    # if shots is None:
    #     shots = 2**12

    # if backend is None:
    #     backend = FakeBrisbane()

    # qc.measure(qr_aux[0], cbit[0])

    # Saída

    return qc
````
