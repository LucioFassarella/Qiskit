## Método de Primitivos



````python
def qiskit_sampler(circuito_com_medidas, shots = None, backend = None):
    '''
    Método para executar o Sampler
    >>> O Sampler realiza medidas em circuitos quânticos e retorna a contagem de qubits.
        A partir dos resultados do Sampler, podemos estimar as probabilidade de cada resultado.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        backend  : backend para rodar o Sampler

    Saída:
        resultado do Sampler

    Observação: os resultados do Sampler são rotulados pelo nomes dos registros clássicos.
                Para um circuito com medidas definidas por 'measure_all', o rótulo é 'meas'.
                Para circuitos com medidas definidas de início sem denominação explícita,
                os rótulos são 'c', 'c1', 'c2', etc.

    Tutorial do Sampler: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/sampler-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import SamplerV2 as Sampler

    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    from qiskit.transpiler import generate_preset_pass_manager
    pm = generate_preset_pass_manager(optimization_level=2, backend = backend)

    # Transpilação

    isa_circuitos = pm.run(circuito_com_medidas)

    if shots is None:
        shots = 2**10
    else:
        shots = shots

    sampler = Sampler(backend)
    job = sampler.run([(isa_circuitos, )], shots = shots)
    result = job.result()

    # Identificação do nome do primeiro registro clássico
    classical_register_name = isa_circuitos.cregs[0].name
    counts = result[0].data[classical_register_name].get_counts()

    return counts, result

def qiskit_estimator(circuito_sem_medidas, hamiltoniano, precisao = None, backend = None):
    '''
    Método para executar o Estimator
    >>> O Estimator estima valores esperados de circuitos e observáveis.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        hamiltoniano (SparsePauliOp): hamiltoniano a ser medido
        precisao (float): precisão do resultado
        backend  : backend para rodar o Estimator

    Saída:
        resultado do Estimator

    Tutorial do EstimatorV2: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/estimator-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import EstimatorV2 as Estimator
    from qiskit.transpiler import generate_preset_pass_manager

    # Use the existing service object
    # service = QiskitRuntimeService()
    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    if precisao is None:
        precisao = 2**(-10)
    else:
        precisao = precisao

    # Transpilação

    pm = generate_preset_pass_manager(backend = backend, optimization_level=1)
    isa_circuito = pm.run(circuito_sem_medidas)
    isa_hamiltoniano = hamiltoniano.apply_layout(isa_circuito.layout)

    # Estimator:

    estimator = Estimator(mode=backend)

    # Job
    job = estimator.run([(isa_circuito, isa_hamiltoniano, )], precision = precisao)
    pub_result = job.result()[0]

    return pub_result.data.evs
````

````python
import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
import qiskit.quantum_info as qi
from qiskit.circuit.library import QFTGate, HamiltonianGate


def quantiza_vetor(b_vetor = list):
    '''
    Algoritmo que codifica um vetor b em amplitude, retornando o circuito que prepara o estado |b > normalizado

    Entrada:
        b_vetor (list): vetor

    Saída:
        qc (QuantumCircuit): circuito quântico que prepara o estado |b > normalizado
    '''

    # Número de qubits para representar o vetor
    num_qubits = int(np.ceil(np.log2(len(b_vetor))))

    # Normalização e completamento do vetor até a coordenada 2**n
    vetor_normalizado = np.array(b_vetor/np.linalg.norm(b_vetor)) # Normaliza o vetor
    vetor_normalizado_completo = np.zeros(2**num_qubits, dtype=complex) # Vetor nulo, ensure complex dtype
    vetor_normalizado_completo[:len(vetor_normalizado)] = vetor_normalizado

    # Circuito de inicialização
    qc = QuantumCircuit(num_qubits)
    qc.initialize(vetor_normalizado_completo, normalize = True)

    return qc

def quantiza_unitario(unitario = list):
    '''
    Algoritmo que quantiza um operador unitário U,
    retornando um circuito que implementa U

    Entrada:
        U (list): operador unitário

    Saída:
        qc (QuantumCircuit): circuito quântico que implementa U
    '''

    dimensao = len(unitario)

    num_qubits = int(np.ceil(np.log2(dimensao)))

    unitario2 = np.identity(2**num_qubits, dtype = complex)

    for i in range(dimensao):
        for j in range(dimensao):
            unitario2[i][j] = unitario[i][j]

    gate = qi.Operator(unitario2)

    qc = QuantumCircuit(num_qubits)

    qc.unitary(gate, range(num_qubits))

    return qc

def quantiza_expoperador(operador_hermitiano = list, tempo = float):
    '''
    Algoritmo que quantiza um operador hermitiano A em dado tempo t,
    retornando o circuito que prepara o unitário exp(-2.pi.i.A.t)

    Entrada:
        operador_hermitiano (list): operador
        time (float)..............: tempo

    Saída:
        qc (QuantumCircuit): circuito quântico que prepara o unitário  exp(-2.pi.i.A.t)
    '''

    dimensao = len(operador_hermitiano)

    num_qubits = int(np.ceil(np.log2(dimensao)))

    hamiltoniano = np.identity(2**num_qubits, dtype = complex)

    for i in range(dimensao):
        for j in range(dimensao):
            hamiltoniano[i][j] = operador_hermitiano[i][j]

    # Cria a porta exp(-2.pi.i.A.t)
    time = 2*np.pi*tempo
    ham_gate = HamiltonianGate(hamiltoniano, time)
    qc = QuantumCircuit(num_qubits)
    qc.append(ham_gate, range(num_qubits))

    return qc

def Crot(num_qubits = int, t0_inverso = None):

    from qiskit.circuit.library import RYGate

    if t0_inverso is None:
        t0_inverso = 2**num_qubits

    # Circuito com n qubits de controle e um qubit auxiliar alvo
    qc = QuantumCircuit(num_qubits + 1)

    # Rotações controladas

    for k in range(1,2**num_qubits):

        # angulo
        bin_string = bin(k)[2:].zfill(num_qubits)
        bin_string_invertida = bin_string[::-1]
        numero = int(bin_string, 2)/2**num_qubits

        seno = 1/(numero*t0_inverso)
        if np.abs(seno) > 1:
            seno = np.sign(seno) # Cap at 1 or -1

        angulo = 2*np.arcsin(seno)

        # rotação controlada

        ry_gate= RYGate(angulo)
        crot = ry_gate.control(num_qubits, ctrl_state = bin_string_invertida)

        # ação

        qc.append(crot, list(range(num_qubits + 1)))

    return qc

def QCEstimativaFase(qc_unitario, qc_autoestado = None, qubits_estado = int, qubits_fase = int):
    '''
    Método que constrói o circuito do algoritmo Estimativa de Fase.

    Inputs:
        qc_unitario (QuantumCircuit)..: circuito quântico que que implementa o operador unitário
        qc_autoestado (QuantumCircuit): circuito quântico que prepara o auto-estado do operador unitário
        qubits_estado (int).......: número de qubits do circuito principal (igual ao número de qubits do qc_autoestado)
        qubits_fase (int).........: número de qubits do circuito auxiliar para armazenar os bits da fase

    Output:
        qc (QuantumCircuit): circuito quântico que implementa o algoritmo Estimativa de Fase.

    Métodos requeridos:
        QuantumCircuit, QuantumRegister, ClassicalRegister
        qiskit.circuit.library.QFTGate
        qiskit_ibm_runtime.SamplerV2
    '''

    # Cria o circuito quântico
    qr_estado = QuantumRegister(qubits_estado, name = 'qr_estado')
    qr_fase = QuantumRegister(qubits_fase, name = 'qr_fase')
    qc = QuantumCircuit(qr_estado, qr_fase)
    qc.name = 'q-QCEF'

    # Preparação do auto-estado
    if qc_autoestado is None:
        pass
    else:
        qc.compose(qc_autoestado, qubits=list(qr_estado), inplace=True)

    # Aplicação de Hadamard nos qubits auxiliares
    for k in range(qubits_fase):
        qc.h(qr_fase[k])

    # Convert qc_unitario to a gate once for efficiency
    unitario_as_gate = qc_unitario.to_gate()

    # Aplicação das potências controladas do operador unitário

    # U^(2^0) = U:
    ctrl_gate_k0 = unitario_as_gate.control(1)

    # Aplicação de U^(2^k)-controlado
    qc.append(ctrl_gate_k0, [qr_fase[0]] + list(qr_estado))

    # U^(2^k), para k = 1,..., num_qubits_aux -1:
    for k in range(1, qubits_fase):
        # Circuito que representa U^(2^k)
        qc_unitario_potencia = QuantumCircuit(qubits_estado)
        for _ in range(2**k):
            qc_unitario_potencia.append(unitario_as_gate, list(range(qubits_estado)))

        # Circuito de U^(2^k)-controlado
        ctrl_gate_potencia = qc_unitario_potencia.control(1)

        # Aplicação de U^(2^k)-controlado
        qc.append(ctrl_gate_potencia, [qr_fase[k]] + list(qr_estado))

    # Aplicação da TFQ-inversa nos qubits auxiliares
    qft_gate = QFTGate(num_qubits = qubits_fase)
    iqft_gate = qft_gate.inverse()
    qc.append(iqft_gate, list(qr_fase))

    return qc
````

````python
def Emax(operador = list):
    '''
    Determina o máximo dos módulos das entradas do operador
    '''
    emax = np.abs(operador[0][0])
    for linha in operador:
        for entrada in linha:
            abs_entrada = np.abs(entrada)
            if abs_entrada > emax:
                emax = abs_entrada

    return emax

def Emin(operador = list):
    '''
    Determina o menor dos módulos das entradas não nulas do operador
    '''
    emin = np.abs(operador[0][0])
    for linha in operador:
        for entrada in linha:
            abs_entrada = np.abs(entrada)
            if abs_entrada > 0 and abs_entrada < emin:
                emin = abs_entrada

    return emin


def QCHHL(qc_expoperador, qc_vetor, qubits_vetor = int, qubits_fase = int, t0 = None):
    '''
    Algoritmo que constrói o circuito HHL que prepara o estado quantizado
    da solução do sistema de equações lineares Ax = b, onde
    - A é um operador hermitiano inversível;
    - b é um vetor.

    Entradas:
        qc_expoperador (QuantumCircuit): circuito que implementa a exponencial do operador hermitiano inversível: exp(-2.pi.i A/t1)
        qc_vetor (QuantumCircuit)......: circuito que prepara a representação quântica do vetor b: |b>
        qubits_vetor (int).: número de qubits do registrar o vetor quantizado (igual ao número de qubits de qc_vetor)
        qubits_fase (int)...: número de qubits do circuito auxiliar para armazenar os bits da fase

    Saída:
        qc_hhl (QuantumCircuit): circuito quântico que implementa o algoritmo HHL, ou seja,
        ........................ constrói o circuito que prepara o estado |x >
        ........................ que é a quantização do vetor A^{-1}b.

    '''

    # Circuito Quântico

    qr_vetor = QuantumRegister(qubits_vetor, name = 'q-vetor')
    qr_fase = QuantumRegister(qubits_fase, name = 'q-fase')
    qr_aux = QuantumRegister(1, name = 'q-aux')
    qc = QuantumCircuit(qr_vetor, qr_fase, qr_aux)
    qbits = qubits_vetor + qubits_fase + 1

    # Inicialização

    quantum_vetor = qc_vetor.copy()
    quantum_vetor.name = 'qc-vetor'

    quantum_expoperador = qc_expoperador.copy()
    quantum_expoperador.name = 'qc-expop'

    qc.compose(quantum_vetor, list(qr_vetor), inplace = True)
    qc.barrier()

    # Estimativa de fase

    # Calculate t0_inverso (1/t0)
    if t0 is None:
        t0_inverso = 2**qubits_fase
    else:
        t0_inverso = 1/t0

    qc_fase = QCEstimativaFase(qc_unitario = quantum_expoperador, qc_autoestado = None, qubits_estado = qubits_vetor, qubits_fase = qubits_fase)
    qc_fase.name = 'qc-fase'

    qc.append(qc_fase, list(qr_vetor) + list(qr_fase))

    # Rotações controladas

    qc_rotacoes = Crot(qubits_fase, t0_inverso = t0_inverso)
    qc_rotacoes.name = 'qc-rotacoes'

    qc.append(qc_rotacoes, list(qr_fase) + list(qr_aux))

    # Estimativa de fase inversa (quantum_vetor = estado quântico preparado na etapa anterior)

    qc_fasei_inversa = qc_fase.inverse()
    qc_fasei_inversa.name = 'qc-fase-inv'

    qc.append(qc_fasei_inversa, list(qr_vetor) + list(qr_fase))

    return qc

def Simulador(qc, aer = True, backend = None, shots = None):

    if shots is None:
        shots = 2**12

    if aer is True:
        from qiskit_aer import AerSimulator
        simulator = AerSimulator()

        # Transpilação

        from qiskit import transpile
        qc_transpilado = transpile(qc, basis_gates=['u', 'cx'], optimization_level=1)

        job = simulator.run(qc_transpilado, shots = shots)
        result = job.result()
        counts = result.get_counts(qc)

    return counts

'''
def Simulador_QCHHL(qc_expoperador, qc_vetor, qubits_vetor = int, qubits_fase = int, t1 = int, t0 = None, aer = True, backend = None, shots = None):

    qr_vetor = QuantumRegister(qubits_vetor)
    qr_fase = QuantumRegister(qubits_fase)
    qr_aux = QuantumRegister(1)
    cr_vetor = ClassicalRegister(qubits_vetor)
    cr_fase = ClassicalRegister(qubits_fase)
    cr_aux = ClassicalRegister(1)

    qbits = qubits_vetor + qubits_fase + 1
    qc = QuantumCircuit(qr_vetor, qr_fase, qr_aux, cr_vetor, cr_fase, cr_aux)

    qc_hhl = QCHHL(operador_hermitiano, vetor, t0 = t0, t1 = t1, qubits_vetor = qubits_vetor, qubits_fase = qubits_fase)

    qc.append(qc_hhl, list(qr_vetor) + list(qr_fase) + list(qr_aux))

    qc.measure(list(qr_vetor), list(cr_vetor))

    qc.measure(list(qr_fase), list(cr_fase))

    qc.measure(list(qr_aux), list(cr_aux))

    if shots is None:
        shots = 2**12

    if aer is True:
        from qiskit_aer import AerSimulator
        simulator = AerSimulator()

        # Transpilação

        from qiskit import transpile
        qc_transpilado = transpile(qc, basis_gates=['u', 'cx'], optimization_level=1)

        job = simulator.run(qc_transpilado, shots = shots)
        result = job.result()
        counts = result.get_counts(qc)

    return counts
'''
````

\vec{u}_1 = \begin{pmatrix}

**Resultado esperado do HHL**

|x\rangle = \alpha |0\rangle + \beta |1\rangle; \

<font color="gray">

\begin{split}

\end{split}

</font>

Parâmetri $t_1$:
$$
t_1 > \max\left(\frac{2}{3},\ \frac{4}{3}\right) \leadsto t_1 = 1.4

Deve valer para todo operador hermitiano $B$ no espaço de Hilbert de $|x\rangle$:
$$
\begin{split}
\langle x |B| x\rangle &= |\alpha|^2 \langle 0 |B|0\rangle + |\beta|^2 \langle 1 |B|1\rangle + \bar{\alpha}\beta \langle 0 |B|1\rangle + \alpha \bar{\beta}\langle 1 |B|0\rangle\\

\end{split}
$$

\begin{array}{c|c|l}

Z &  |\alpha|^2 - |\beta|^2 & = - \frac{8}{10}\\

X & 2\Re(\alpha \bar{\beta}) & = \frac{6}{10}\\
