## Método de Primitivos



```python
def qiskit_sampler(circuito_com_medidas, shots = None, backend = None):
    '''
    Método para executar o Sampler
    >>> O Sampler realiza medidas em circuitos quânticos e retorna a contagem de qubits.
        A partir dos resultados do Sampler, podemos estimar as probabilidade de cada resultado.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        backend  : backend para rodar o Sampler

    Saída:
        resultado do Sampler

    Observação: os resultados do Sampler são rotulados pelo nomes dos registros clássicos.
                Para um circuito com medidas definidas por 'measure_all', o rótulo é 'meas'.
                Para circuitos com medidas definidas de início sem denominação explícita,
                os rótulos são 'c', 'c1', 'c2', etc.

    Tutorial do Sampler: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/sampler-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import SamplerV2 as Sampler

    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    from qiskit.transpiler import generate_preset_pass_manager
    pm = generate_preset_pass_manager(optimization_level=2, backend = backend)

    # Transpilação

    isa_circuitos = pm.run(circuito_com_medidas)

    if shots is None:
        shots = 2**10
    else:
        shots = shots

    sampler = Sampler(backend)
    job = sampler.run([(isa_circuitos, )], shots = shots)
    result = job.result()

    # Identificação do nome do primeiro registro clássico
    classical_register_name = isa_circuitos.cregs[0].name
    counts = result[0].data[classical_register_name].get_counts()

    return counts, result

def qiskit_estimator(circuito_sem_medidas, hamiltoniano, precisao = None, backend = None):
    '''
    Método para executar o Estimator
    >>> O Estimator estima valores esperados de circuitos e observáveis.

    Input:
        circuito : circuito quântico que prepara um estado de qubits
        hamiltoniano (SparsePauliOp): hamiltoniano a ser medido
        precisao (float): precisão do resultado
        backend  : backend para rodar o Estimator

    Saída:
        resultado do Estimator

    Tutorial do EstimatorV2: https://quantum.cloud.ibm.com/docs/en/api/qiskit-ibm-runtime/estimator-v2
    '''

    # Importação de métodos e backend

    from qiskit_ibm_runtime import EstimatorV2 as Estimator
    from qiskit.transpiler import generate_preset_pass_manager

    # Use the existing service object
    # service = QiskitRuntimeService()
    if backend is None:
        from qiskit_ibm_runtime.fake_provider import FakeBrisbane
        backend = FakeBrisbane()
    else:
        backend = backend

    if precisao is None:
        precisao = 2**(-10)
    else:
        precisao = precisao

    # Transpilação

    pm = generate_preset_pass_manager(backend = backend, optimization_level=1)
    isa_circuito = pm.run(circuito_sem_medidas)
    isa_hamiltoniano = hamiltoniano.apply_layout(isa_circuito.layout)

    # Estimator:

    estimator = Estimator(mode=backend)

    # Job
    job = estimator.run([(isa_circuito, isa_hamiltoniano, )], precision = precisao)
    pub_result = job.result()[0]

    return pub_result.data.evs
```

# Estado Fundamental via VQE

Protocolo:

1.  Definição do operador parametrizado de evolução discreta:
$$
\mathbb{R}^m \ni \vec{\lambda} \mapsto U(\vec{\lambda}) \in \mathcal{U}(\mathbb{Q}_n)
$$
2.  Definição/preparação do estado inicial:
$$
| 0 \rangle \xrightarrow{U_0} | \psi_0\rangle
$$
3.  Inicialização do parâmetro de evolução:
$$
\vec{\lambda}_0 \in \mathbb{R}^m
$$
4.  Estimativa da energia inicial:
$$
E_0 = \langle \psi_0 | H | \psi_0 \rangle
$$
5. Iteração de $k = 0$ até $k = \ell$:
    *   Atuação do operador de evolução:
$$
| \psi_k \rangle \xrightarrow{U(\vec{\lambda_k})} | \psi_{k+1} \rangle
$$
    *   Medição do valor esperado do Hamiltoniano:
$$
E_{k+1} = \langle \psi_{k+1} | H | \psi_{k+1} \rangle
$$

## Exemplo 1: Sistema Unidimensional

Sitema: 1 qubit,
$$
\mathcal{H} = \mathcal{Q}
$$

Hamiltoniano
$$
H = Z
$$

Estado inicial (~ circuito de preparação):
$$
| \psi_0 \rangle \equiv \mathrm{H}| 0 \rangle = \frac{| 0 \rangle + | 1 \rangle}{\sqrt{2}}
$$
Operador de evolução (~ circuito de evolução parametrizado):
$$
u(\theta, \lambda, \phi) = \begin{pmatrix}
\cos\left(\frac{\theta}{2}\right) & - e^{-i\lambda}\sin\left(\frac{\theta}{2}\right)\\
e^{-i\phi}\sin\left(\frac{\theta}{2}\right) & e^{i(\phi + \lambda)}\cos\left(\frac{\theta}{2}\right)
\end{pmatrix}
$$
Método de otimização: sympy.minimize(Cobyla):
$$
scipy.optimize.minimize
$$

[https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize)


```python
'''
PROTOCOLO VQE

Elementos requeridos:
    > Hamiltoniano
    > Circuito de preparação do estado inicial
    > Circuito de evolução parametrizado
    > Parâmetros iniciais
'''

#Métodos requeridos:
import numpy as np
from qiskit.circuit import QuantumCircuit, Parameter, ParameterVector
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit.library import PauliEvolutionGate
from scipy.optimize import minimize

# Número de qubits
num_qubits = 1

# Hamiltoniano
hamiltoniano = SparsePauliOp(["I", "X", "Y", "Z"], coeffs=[0, 0, 0, 1])
# >>> Impressão:
gate = PauliEvolutionGate(hamiltoniano)
print(gate.to_matrix())

# Circuito de preparação do estado inicial
qc_preparacao = QuantumCircuit(num_qubits)
for k in range(num_qubits):
    qc_preparacao.h(k)

print("Circuito de Preparação:")
display(qc_preparacao.draw('mpl'))

# Circuito de evolução parametrizado
num_parametros = 3
parametros = ParameterVector('parâmetros', num_parametros)

qc_evolucao = QuantumCircuit(num_qubits)
qc_evolucao.u(parametros[0], parametros[1], parametros[2], 0)

print("Circuito de Evolução:")
display(qc_evolucao.draw('mpl'))

# Parâmetros iniciais (list of numbers)
parametros_iniciais = [0,0,0]
print(f"Parâmetros iniciais: {parametros_iniciais}")

# Função valor esperado
precisao = 2**(-10)
from qiskit_ibm_runtime.fake_provider import FakeBrisbane
backend = FakeBrisbane()

def ExpVal(params):

    qc = QuantumCircuit(num_qubits)

    qc.append(qc_preparacao, range(num_qubits))

    qc_evolucao_instance = qc_evolucao.copy()
    qc.append(qc_evolucao_instance, range(num_qubits))

    qc.assign_parameters({parametros: params}, inplace=True)

    expval = qiskit_estimator(qc, hamiltoniano, precisao = precisao, backend = backend)

    return expval

# Otimização
# >>> Ponto inicial
x0 = np.array(parametros_iniciais.copy())
result = minimize(ExpVal, x0, method = 'COBYLA') # <-- Método de otimização
result
```

```python
# Valor mínimo:
valor_minimo = result.fun
print(f"Valor mínimo: {valor_minimo}")

# Parâmetros ótimos:
parametros_otimo = result.x
print(f"Parâmetros ótimos: {parametros_otimo}")

# Estado fundamental:
qc = QuantumCircuit(num_qubits) # <-- criação do circuito de preparação
qc.append(qc_preparacao, range(num_qubits))
qc_evolucao_otima = qc_evolucao.copy()
qc_evolucao_otima.assign_parameters({parametros: parametros_otimo}, inplace=True)
qc.append(qc_evolucao_otima, range(num_qubits))
display(qc.draw('mpl'))

from qiskit.quantum_info import Statevector
estado_fundamental = Statevector.from_instruction(qc) # <-- visualização do estado preparado
print("Estado fundamental calculado:")
print(estado_fundamental.draw())
print(estado_fundamental.draw('latex_source'))
```

```python
'''
Cálculo do erro relativo entre o estado fundamental exato e o estado fundamental calculado
'''
estado_fundamental_exato = np.array([0,1])
estado_fundamental_calculado = np.array([estado_fundamental[0], estado_fundamental[1]])
produto_interno = np.dot(estado_fundamental_calculado, estado_fundamental_exato)
componente_paralela = produto_interno * estado_fundamental_exato
norma_componente_paralela = np.linalg.norm(componente_paralela)
componente_ortogonal = estado_fundamental_calculado - produto_interno * estado_fundamental_exato
norma_componente_ortogonal = np.linalg.norm(componente_ortogonal)
print(f"Erro ralativo (norma da componente ortogonal dividida pela norma da componente paralela): {(norma_componente_ortogonal/norma_componente_paralela)* 100} %")
```
